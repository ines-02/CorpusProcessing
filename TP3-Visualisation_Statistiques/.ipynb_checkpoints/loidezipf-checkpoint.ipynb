```python
"""
Analyse de la Loi de Zipf sur un corpus de synopsis de films

Ce script utilise spaCy pour tokenizer un corpus de synopsis de films en franais,
compte la frquence des mots, et trace une courbe log-log pour vrifier la distribution
de Zipf dans les donnes textuelles.

Fonctionnalits :
-----------------
1. Chargement de tous les fichiers `.txt` dun dossier corpus (`../film_synopses`).
2. Tokenisation avec `spaCy` (modle `fr_core_news_sm`).
3. Nettoyage : suppression de la ponctuation et des espaces.
4. Comptage de la frquence de chaque token.
5. Tri des frquences de mots par ordre dcroissant.
6. Affichage dune courbe en chelle log-log (Loi de Zipf) :
   - Axe X : rang du mot (ordre de frquence dcroissante)
   - Axe Y : frquence du mot

Pr-requis :
------------
- Python 3
- Bibliothques : `spacy`, `matplotlib`, `collections`
- Modle spaCy franais (`python -m spacy download fr_core_news_sm`)
- Dossier `../film_synopses` contenant des fichiers `.txt`

Utilit :
---------
- Vrifier si la distribution des mots respecte la loi de Zipf, typique des langues naturelles.
- Aide  lanalyse linguistique du corpus, pour NLP, classification ou gnration.

Rsultat :
----------
- Une courbe log-log qui doit approximer une ligne droite si la loi de Zipf est respecte.
- Permet didentifier les mots dominants ou les dsquilibres lexicaux dans les donnes.

"""

import os
from collections import Counter
import spacy
import matplotlib.pyplot as plt

# Charger spaCy franais
nlp = spacy.load('fr_core_news_sm')

corpus_folder = '../film_synopses'

concat_context = []

# Charger tous les fichiers texte du dossier
for filename in os.listdir(corpus_folder):
    if filename.endswith('.txt'):
        with open(os.path.join(corpus_folder, filename), 'r', encoding='utf-8') as f:
            text = f.read()
            doc = nlp(text)
            for token in doc:
                # Filtrer ponctuations et espaces
                if not token.is_punct and not token.is_space:
                    concat_context.append(token.text.lower())

# Compter la frquence des tokens
concat_context_counter = Counter(concat_context)

# Trier frquences en ordre dcroissant
freq_sorted = sorted(concat_context_counter.values(), reverse=True)

# Afficher la courbe log-log de Zipf
plt.figure(figsize=(10, 6))
plt.loglog(range(1, len(freq_sorted)+1), freq_sorted, marker='.')
plt.title("Loi de Zipf - Frquence des mots dans le corpus")
plt.xlabel("Rang du mot (log scale)")
plt.ylabel("Frquence du mot (log scale)")
plt.grid(True, which='both', ls='--', lw=0.5)
plt.show()


```
