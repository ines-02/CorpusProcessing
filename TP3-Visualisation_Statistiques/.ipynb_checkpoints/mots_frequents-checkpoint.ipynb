```python
"""
Script d'analyse lexicale- mots fréquents

Ce script réalise les étapes suivantes :
1. Charge tous les fichiers texte (.txt) contenant les synopsis de films d'horreur.
2. Concatène et nettoie le texte en supprimant la ponctuation et les caractères non alphabétiques.
3. Compte la fréquence d'apparition d'une liste enrichie de mots-clés typiques du genre horreur, 
   répartis en plusieurs catégories (thèmes surnaturels, violence, ambiance, lieux, créatures, etc.).
4. Affiche ces fréquences dans la console, triées par ordre décroissant.
5. Visualise les 30 mots-clés les plus fréquents sous forme d'un histogramme.
6. Génère et affiche un nuage de mots représentant la fréquence des mots-clés dans le corpus.

L'objectif est d'explorer la distribution des éléments d'horreur dans les résumés, 
afin d'aider à la conception ultérieure d'un modèle NLP pour la classification et l'extraction
des éléments typiques du genre horreur.
"""
import os
import re
import matplotlib.pyplot as plt
from collections import Counter
from wordcloud import WordCloud

# Chemin vers le dossier contenant les synopsis
corpus_folder = '../film_synopses' 

# Liste enrichie de mots-clés d'horreur
keywords = [
    # Thèmes surnaturels
    "esprit", "fantôme", "démon", "possédé", "malédiction", "sorcellerie", "rituel", "exorcisme", "enfer", "pacte", "apocalypse",

    # Violence et gore
    "tueur", "meurtre", "massacre", "sacrifice", "torture", "sang", "cadavre", "hurlement", "agonie", "déchiqueté", "mutilation",

    # Peur, ambiance
    "peur", "terreur", "cauchemar", "angoisse", "panique", "crie", "épouvante", "frisson", "folie", "paranoïa", "hallucination",

    # Lieux et objets
    "maison", "manoir", "hôpital", "asile", "forêt", "cimetière", "grenier", "cave", "miroir", "poupée", "croix", "talisman",

    # Créatures
    "monstre", "zombie", "vampire", "loup-garou", "momie", "clown", "chasseur", "entité", "ombre", "spectre", "revenant",

    # Innocence corrompue
    "enfant", "bébé", "innocent", "orphelin", "famille", "adoption"
]

#  Chargement de tout le corpus
all_text = ""
for filename in os.listdir(corpus_folder):
    if filename.endswith(".txt"):
        with open(os.path.join(corpus_folder, filename), 'r', encoding='utf-8') as f:
            all_text += f.read().lower() + " "

# Nettoyage du texte
all_text = re.sub(r'[^a-zàâçéèêëîïôûùüÿñæœ\s]', '', all_text)

# Compter les occurrences des mots-clés
keyword_freqs = {word: all_text.count(word) for word in keywords}

# Trier les mots par fréquence décroissante
sorted_keywords = sorted(keyword_freqs.items(), key=lambda x: x[1], reverse=True)

# Affichage console
print("\nFréquences des mots-clés d'horreur :")
for word, count in sorted_keywords:
    print(f"{word} : {count}")

# Graphe
words_plot, freqs_plot = zip(*sorted_keywords[:30])  # Top 30

plt.figure(figsize=(12,6))
plt.bar(words_plot, freqs_plot)
plt.title("Top 30 mots-clés d'horreur les plus fréquents")
plt.xticks(rotation=45)
plt.ylabel("Occurrences")
plt.tight_layout()
plt.show()

# Nuage de mots
wordcloud = WordCloud(width=800, height=400, background_color='black', colormap='Reds').generate_from_frequencies(keyword_freqs)

plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Nuage de mots – Éléments d'horreur")
plt.show()
```
