```python
"""
Script d'analyse de la longueur des synopsis de films (en nombre de mots)

Ce script lit tous les fichiers texte contenant des synopsis de films dans un dossier donnÃ© 
et calcule des statistiques sur leur longueur, mesurÃ©e en nombre de mots. Il gÃ©nÃ¨re Ã©galement 
un graphique en barres pour visualiser la rÃ©partition des longueurs par film.

FonctionnalitÃ©s principales :
-----------------------------
1. Lecture de chaque fichier `.txt` depuis un dossier de corpus (dÃ©fini par `corpus_folder`).
2. Comptage du nombre de mots dans chaque synopsis.
3. Calcul de statistiques globales :
   - Longueur moyenne des synopsis
   - Ã‰cart-type des longueurs
4. Affichage dâ€™un graphique en barres :
   - Axe X : titres des films
   - Axe Y : nombre de mots par synopsis

PrÃ©-requis :
------------
- Python 3
- BibliothÃ¨ques : `matplotlib`, `numpy` (`pip install matplotlib numpy`)
- Un dossier `../film_synopses` contenant les fichiers `.txt` (un par film)

Utilisation :
-------------
- Ce script permet de dÃ©tecter les Ã©ventuelles anomalies (fichiers trop courts ou trop longs),
  et dâ€™analyser la cohÃ©rence du corpus avant traitement NLP.

RÃ©sultat :
----------
- Affiche un graphique montrant la longueur des synopsis.
- Affiche la longueur moyenne et lâ€™Ã©cart-type des textes analysÃ©s.
"""


import os
import matplotlib.pyplot as plt
import numpy as np

# Dossier oÃ¹ se trouvent les synopsis
corpus_folder = '../film_synopses'

lengths = []       # Liste des longueurs
file_names = []    # Noms des films

# Parcourir tous les fichiers texte dans le dossier
for filename in os.listdir(corpus_folder):
    if filename.endswith(".txt"):
        with open(os.path.join(corpus_folder, filename), 'r', encoding='utf-8') as f:
            text = f.read()
            lengths.append(len(text.split()))               # Compter les mots
            file_names.append(filename.replace('.txt', '')) # Nettoyer le nom du fichier

# Statistiques de base
print("Longueur moyenne :", np.mean(lengths))
print("Ã‰cart-type :", np.std(lengths))

# Affichage graphique
plt.figure(figsize=(12,6))
plt.bar(file_names, lengths)
plt.xticks(rotation=90)
plt.title("Longueur des synopsis (en nombre de mots)")
plt.xlabel("Film")
plt.ylabel("Nombre de mots")
plt.tight_layout()
plt.show()

```
